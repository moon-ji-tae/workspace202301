# 블로터의 실시간 랭킹 

- https://www.bloter.net/
- 순위, 제목, URL 

```python
# 1) 관련 모듈 임포트 
import requests
from bs4 import BeautifulSoup
```

```python
# 2) 웹페이지 주소 요청 => 문자열 데이타 => 숩객체 
# User-Agent 정보 확인  
# https://www.useragentstring.com/
user_info = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
headers_data = {'User-Agent':user_info}
url = 'https://www.bloter.net/'
res = requests.get(url, headers=headers_data)
html_str = res.text
soup = BeautifulSoup(html_str)
# print(soup)
```

```python
# 3) 파싱 
# 관련 소스 확인 
```

```python
# 각각의 a 태그 
target_list = soup.select('.box.main_ranknews > .list > .article > .text_box > .link')
print(len(target_list))
print(target_list[0])
print()
print(target_list[-1])
```

```python
# 제목, url 
base_url = 'https://www.bloter.net'
rank = 1
for a in target_list:
    print(f'{rank}위 {a.h4.text}')
    print(base_url+a['href'])
    rank += 1
    print()
```

```python
# 4) 리스트 구조에 저장하기 
# 순위, 제목, url 
# 리스트안의 리스트 형태 
base_url = 'https://www.bloter.net'
rank = 1
result_list = []
for a in target_list:
    # 리스트 형태로 추가 
    result_list.append([str(rank)+'위', a.h4.text, base_url+a['href']])
    rank += 1
    
print(result_list[0]) 
# print(result_list) 
```

```python
for rank, title, url in result_list:
    print(rank, title, url)
    print()
```

```python
# 5) 텍스트 파일에 저장하기 
with open('output/블로터실시간랭킹.txt', 'w', encoding='utf-8') as f:
    f.write('순위, 제목, URL\n')
    for rank, title, url in result_list:
        f.write(f'{rank},{title},{url}\n')
```

```python
# 6) 판다스의 테이블 구조로 변경하기 
# 판다스란? 파이썬계의 엑셀 
# 모듈 임포트 
import pandas as pd
# pd.DataFrame(2차원리스트|리스트안의딕셔너리, 
#               columns=[컬럼명 .... ]  )
df = pd.DataFrame(result_list, 
                  columns=['순위', '제목', 'URL'])
df
```

```python
# 7) csv 파일로 저장하기 
# 쉼표를 이용해서 데이타를 분리시켜 저장 
# 판다스 데이타프레임 => csv
# 데이타프레임.to_csv(파일url)
# encoding='cp949' 엑셀에서 한글깨짐 방지
df.to_csv('output/블로터랭킹.csv', encoding='cp949')
```
