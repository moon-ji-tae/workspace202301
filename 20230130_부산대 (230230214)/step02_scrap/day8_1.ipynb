      : 선택자1 선택자2 ... { 속성1:값1;속성2:값2;... }

    - 구조선택자 : <br>
        선택자:first-child { 속성1:값1;속성2:값2;... }
        선택자:last-child { 속성1:값1;속성2:값2;... }

### 웹페이지요청 => 객체 => 바이트 데이타 => 문자열 =>  html 파일

```python
# 1) 페이지 요청 => response 객체 
url = 'https://www.daum.net/'
res = urlopen(url)
# 응답코드 확인 
print(res.status, res.code)

# 2) response 객체 => html 소스(바이트) => 문자열 
# decode(인코딩) : 바이트 => 문자열
html_str = res.read()
html_str = html_str.decode('utf-8')
print(type(html_str))

# 3) html 소스(문자열) => html 파일 페이지 저장 
with open('output/daum2.html', 'w', encoding='utf-8') as f:
    f.write(html_str)
```

### 블로터 페이지 저장 

웹페이지요청(헤더정보) => 객체 => 바이트 데이타 => 문자열 =>  html 파일

```python
# 0) urllib.requests 모듈의 urlopen, Request 함수만 임포트 
from urllib.request import urlopen, Request
```

```python
# 1) Request 객체 생성 
# a) 웹브라우저 headers 정보 확인 
# https://www.useragentstring.com/
# b) headers 딕셔너리 만들기 
user_info = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
headers_data = {'User-Agent':user_info}
url = 'https://www.bloter.net/'
req = Request(url, headers=headers_data)
print(req)
```

```python
# 2) Request 객체 => response 객체 
res = urlopen(req)
# 응답코드 확인 
print(res.status, res.code)

# 3) response 객체 => html 소스(바이트) => 문자열 
# decode(인코딩) : 바이트 => 문자열
html_str = res.read()
html_str = html_str.decode('utf-8')
print(type(html_str))

# 4) html 소스(문자열) => html 파일 페이지 저장 
with open('output/블로터.html', 'w', encoding='utf-8') as f:
    f.write(html_str)
```

## 웹페이지 => 문자열 => 숩객체화 

```python
import requests

# 국립고궁박물관 페이지 
url = 'https://www.gogung.go.kr/gogung/main/contents.do?menuNo=800011'
res = requests.get(url)
print(res)
html_str = res.text
print(type(html_str))

# 문자열데이타 => 숩객체 
# BeautifulSoup(html데이타, 'html.parser | xml')
soup = BeautifulSoup(html_str, 'html.parser')
print(type(soup))
# print(soup)
```

# find() , find_all()

```python
# 1) html 소스 요청 => html 문자열 => 숩객체
url = 'https://www.gogung.go.kr/gogung/main/contents.do?menuNo=800011'
res = requests.get(url)
html_str = res.text
soup = BeautifulSoup(html_str, 'html.parser')
```

```python
# 2) 파싱 테스트 
# b 태그 찾기 
# 첫번째 b 태그 요소 
print(soup.find('b'),type(soup.find('b')))
print(soup.find('b').text)
print()
# 페이지에 삽입된 모든 b 태그요소
print(type(soup.find_all('b')))
print(soup.find_all('b'))

# 리스트 형태로 변경 
b_tag_list = list(soup.find_all('b'))
print()
print(type(b_tag_list), len(b_tag_list))

# b 태그안의 내용물만 출력하기 
print()
for word in soup.find_all('b'):
    print(word.text)
    
# 마지막에 삽입된 b 태그의 텍스트 출력 
print()
print(soup.find_all('b')[-1].text)
```

```python
# img 태그 찾기 
img_list = soup.find_all('img')
print('img 태그 총갯수는?', len(img_list))
# print(img_list)
# img 태그에서 alt, src 속성값 출력하기 
base_url = 'https://www.gogung.go.kr'
for img in img_list:
    print('='*10)
    print(img['alt'])
    print(base_url+img['src'])
    
```

```python
# id 값으로 찾기 
# find(id=아이디명)
target = soup.find(id='menu')
# 태그명 확인 
print(target.name)
print()
print(target)
```

```python
# nav 태그안의 첫번째 p 태그 찾기
print(target.p)
print(target.p.text)

# nav 태그안의 모든 p 태그 찾기 
p_list = target.find_all('p')
# print(p_list)
print('='*10)
for p in p_list:
    print(p.text.strip())
    print('*'*20)
```

```python
# 클래스값으로 찾기 
# find(class_= 클래스명)
# find_all(class_= 클래스명)
target = soup.find(class_='title')
print(target)
target_list = soup.find_all(class_='title')
print(len(target_list))
```

```python
# 관람시 유의사항 텍스트 부분만 추출하기 
target = soup.find(class_='num-list')
# print(target)
# num-list 클래스 값을 가진 태그요소 안에서 
# 다시 클래스 값이 text인 요소 찾기 
span_list = target.find_all(class_='text')
print(len(span_list))
# 텍스트만 출력하기 
count = 1
for span in span_list:
    print()
    print(count, span.text)  
    count += 1
```

```python
# 속성값으로 찾기 
# find(attrs={속성:값}) => 태그 1개
# find_all(attrs={속성:값}) => 태그 여러개 
```

```python
# href 속성값이 # 인 태그 요소찾기 
target_list = soup.find_all(attrs={'href':'#'})
print(len(target_list))
print(target_list[0])
print(target_list[-1])
```

```python
# 하단메뉴의 메뉴명과 하이퍼링크주소 추출하기 
# clss 값이 btns 인 요소 찾기 
# target = soup.find(class_ = 'btns')
target = soup.find(attrs={'class':'btns'})
print(target)
```

```python
# 메뉴명과 주소 출력하기 
# 첫번째 메뉴만 출력 
print(target.a.text)
print(target.a['href'])
```

```python
# 모든 메뉴 출력하기 
a_list = target.find_all('a')
for a in a_list:
    print(a.text)
    print(a['href'])
    print('='*10)
```

```python
# 모든 메뉴 출력하기 
# 페이지 주소 완성 
a_list = target.find_all('a')
base_url = 'https://www.gogung.go.kr'
for a in a_list:
    print(a.text)
    if a['href'][0] == '/':
        print(base_url+a['href'])
    else:
        print(a['href'])
    print('='*10)
```
