```python
# 모듈 임포트 
import requests

# print(dir(requests))
```

```python
# 특정 웹페이지 요청 => response 객체 생성. get()
# => html 소스 (텍스트) => html 파일 저장 
url = 'http://google.com'
res = requests.get(url)
# 200 확인
print(res)
# 인코딩 정보 
print(res.encoding)
# 요청한 url 페이지의 정보 확인 
print(res.headers)
```

```python
# 웹페이지소스 => 파이썬의 문자열 데이타 
html_str = res.text
print(type(html_str))
# print(html_str)
```

```python
# 웹페이지 문자열 데이타 => html 파일 저장
with open('output/google.html', 'w') as f:
    f.write(html_str)
```

```python
ls output\*.html
```

```python
# yes24 시작페이지 저장하기 
url = 'http://yes24.com'
res = requests.get(url)
# 200 확인
print(res)
# 인코딩 정보 확인
print(res.encoding)
# 문자열 데이타로 저장 
html_str = res.text
# print(html_str)
# 웹페이지 문자열 데이타 => html 파일 저장
# encoding 속성과 값 전달 
with open('output/yes24.html', 'w', encoding='utf-8') as f:
    f.write(html_str)
```

```python
ls output\*.html
```

```python
# 블로터 시작 페이지 저장 
# headers 정보 만들어서 웹페이지 요청하기  
url = 'https://www.bloter.net/'
res = requests.get(url)
print(res)
# 문자열 데이타로 저장 
html_str = res.text
print(html_str)
```

```python
# 1) 웹브라우저 headers 정보 확인 
# https://www.useragentstring.com/
# 2) headers 딕셔너리 만들기 
user_info = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
headers_data = {'User-Agent':user_info}
print(headers_data)
# 3) 페이지 요청 => response 객체 
url = 'https://www.bloter.net/'
res = requests.get(url, headers=headers_data)
print(res)
# 4) 문자열 데이타로 저장 
html_str = res.text
print(html_str)
```
