```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
# 발급받은 클라이언트아이디와 시크릿키를 문자열키로 지정
client_id = "J78os7A0HOCZCrDxnwLy"
client_secret = "ZYoNiYyUrZ"
# 샘플 키워드 설정 
encText = urllib.parse.quote("부산대맛집")
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# 20 개 요청
url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=20"  # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
# json => 문자열 => 파이썬 데이타 (딕셔너리 => 리스트)
import json 

json_str = response_body.decode('utf-8')
print(type(json_str))
blog_dict = json.loads(json_str)
print(type(blog_dict))
# 실제 블로그 데이타를 키값으로 호출한후 저장 
blog_list = blog_dict['items']
print(type(blog_list), len(blog_list))
# print(blog_list[0])
```

### 3) 테이블 구조 => 데이타프레임 => csv  파일로 저장

```python
import pandas as pd 
```

```python
# 개별 블로그 데이타 1개
# <b>와 </b> 삭제 
title = blog_list[0]['title'].replace('<b>','').replace('</b>','')
link = blog_list[0]['link']
description = blog_list[0]['description'].replace('<b>','').replace('</b>','')
bloggername = blog_list[0]['bloggername']
postdate = blog_list[0]['postdate']
print('제목 : ', title)
print('블로그 페이지 :', link)
print('내용 :', description)
blog_list[0]
print('작성자 :', bloggername)
print('작성날짜 :', postdate)
```

```python
# 테이블 구조 - 리스트안의 리스트
result_list = []
for blog in blog_list:
    title = blog['title'].replace('<b>','').replace('</b>','').replace('\u2764','').replace('\ufe0f','').replace('\u02f5','').replace('\u25d5','')
    link = blog['link'].replace('\u2764','').replace('\ufe0f','').replace('\u02f5','').replace('\u25d5','')
    description = blog['description'].replace('<b>','').replace('</b>','').replace('\u2764','').replace('\ufe0f','').replace('\u02f5','').replace('\u25d5','')
    bloggername = blog['bloggername'].replace('\u2764','').replace('\ufe0f','').replace('\u02f5','').replace('\u25d5','')
    postdate = blog['postdate'].replace('\u2764','').replace('\ufe0f','').replace('\u02f5','').replace('\u25d5','')    
    result_list.append([title, link, description, bloggername,postdate ])
```

```python
print(len(result_list))
```

```python
# 데이타프레임으로 변경 
df = pd.DataFrame(result_list, 
       columns = ['제목', '링크', '설명', '작성자', '작성날짜'])
# df
```

```python
# csv 파일로 저장 
# df.to_csv('output/네이버블로그검색.csv', index=False)
df.to_csv('output/네이버블로그검색.csv', index=False, encoding='cp949')
```

# 네이버 지식인 검색 서비스 

- https://openapi.naver.com/v1/search/kin.json

```python
# 0) 모듈 임포트 
import os
import sys
import urllib.request
import json
import pandas as pd 
```

```python
# 1) 지식인 검색 데이타 요청 
# 발급받은 클라이언트아이디와 시크릿키를 문자열키로 지정
client_id = "J78os7A0HOCZCrDxnwLy"
client_secret = "ZYoNiYyUrZ"
# 샘플 키워드 설정 
encText = urllib.parse.quote("부산대맛집")
# 지식인 주소 확인 : kin.json
# 20 개 요청
url = "https://openapi.naver.com/v1/search/kin.json?query=" + encText + "&display=20"  # JSON 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
# 2) json => 문자열 => 파이썬 데이타 (딕셔너리 => 리스트)
import json 

json_str = response_body.decode('utf-8')
print(type(json_str))
kin_dict = json.loads(json_str)
print(type(kin_dict))
# 실제 블로그 데이타를 키값으로 호출한후 저장 
kin_list = kin_dict['items']
print(type(kin_list), len(kin_list))
print(kin_list[0])
```

```python
# 3) 테이블 구조 => 데이타프레임 => csv 파일로 저장
```

```python
# 개별 지식인 검색 결과 데이타 1개
# <b>와 </b> 삭제 
title = kin_list[0]['title'].replace('<b>','').replace('</b>','')
link = kin_list[0]['link']
description = kin_list[0]['description'].replace('<b>','').replace('</b>','')
print('제목 : ', title)
print('페이지 :', link)
print('내용 :', description)
```

```python
# 테이블 구조 - 리스트안의 리스트
result_list = []
for kin in kin_list:
    title = kin['title'].replace('<b>','').replace('</b>','')
    link = kin['link']
    description = kin['description'].replace('<b>','').replace('</b>','')
    result_list.append([title, link, description])
```

```python
# 데이타프레임으로 변경 
df = pd.DataFrame(result_list, 
       columns = ['제목', '링크', '설명'])
df.head(2)
```

```python
# csv 파일로 저장 
df.to_csv('output/네이버지식인검색.csv', index=False, encoding='cp949')
```

### 전체 측정소 데이타 

```python
# 1) 데이타 요청 

# 서비스키 변수 설정 
myKey = "Yhq6n9vlpaKZsCOw0jec8zIdZ8p+Bpuku6WLjVgDhqXRW6dHnfXoauSEj19jpnjv59CUGspyTxVCgCeXOua7dg=="
url = 'http://apis.data.go.kr/1480523/MetalMeasuringResultService/MetalService'

# 리스트 초기화 
dust_list = []

for i in range(1,11):
    params ={'serviceKey' : myKey, # 키변수값으로변경
             'pageNo' : '1', 
             'numOfRows' : '12', 
             'resultType' : 'JSON', # 데이타타입
             'date' : '20230101', # 날짜
             'stationcode' : str(i), # 문자열로 변경 '1'~'10'
             'itemcode' : '90303',  # 납 90303 칼슘 90319
             'timecode' : 'RH02' } # RH02, RH24

    print(i, end=' ')
    try:
        response = requests.get(url, params=params)    
        dust = json.loads(response.text)
        dust_list += dust['MetalService']['item']
    except:
        pass
```

```python
result_list = []
for item in dust_list:
    temp = station[str(item['STATIONCODE'])]
    result_list.append([item['SDATE'], temp, item['VALUE']])
```

```python
df = pd.DataFrame(result_list, 
                  columns=['날짜시간', '대기환경연구소', '측정수치'])
df.to_csv('output/미세먼지.csv', 
          index=False, 
          encoding='cp949' )
df.head(15)
```
